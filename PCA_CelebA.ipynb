```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
import sklearn.preprocessing as preprocessing
```

```python
data_root = "./smallerdata/"
all_images = os.listdir(data_root)
print("Total number of images:", len(all_images) )
```

## Sample images and merge into dataset

```python
number_of_samples = 2000
np.random.seed(seed=1)
sample_index = np.random.choice(len(all_images),number_of_samples)
sample_images = np.array(all_images)[sample_index]

PCA_data = []
for i in sample_images:
    this_image = list(np.asarray(Image.open(data_root+i)).flatten())
    PCA_data.append(this_image)
PCA_data = np.array(PCA_data)

```

```python
# Add randy's protrait to selected iamge
this_image = np.asarray(Image.open(data_root+"randy.jpg"))
from skimage.transform import resize
resized_image = resize(this_image, (218, 178, 3))* 255
plt.imshow((resized_image).astype(int),
              cmap = plt.cm.gray, interpolation='nearest',
              clim=(0, 255));
plt.xlabel("All channels", fontsize = 12)
plt.title('Original Image', fontsize = 13);
PCA_data[0] = list(resized_image.flatten())
```

```python
# fit multiple pca with different percentage of explained Variance

pca6 = PCA(.6)
lower_dimensional_data = pca6.fit_transform(PCA_data)
n_components6 = len(pca6.explained_variance_ratio_)
approximation6 = pca6.inverse_transform(lower_dimensional_data)

pca8 = PCA(.8)
lower_dimensional_data = pca8.fit_transform(PCA_data)
n_components8 = len(pca8.explained_variance_ratio_)
approximation8 = pca8.inverse_transform(lower_dimensional_data)

pca95 = PCA(.95)
lower_dimensional_data = pca95.fit_transform(PCA_data)
n_components95 = len(pca95.explained_variance_ratio_)
approximation95 = pca95.inverse_transform(lower_dimensional_data)

pca99 = PCA(.99)
lower_dimensional_data = pca99.fit_transform(PCA_data)
n_components99 = len(pca99.explained_variance_ratio_)
approximation99 = pca99.inverse_transform(lower_dimensional_data)
```

## Plot the decay of the singular values

```python

plt.figure(figsize=(10,7))
plt.plot(pca8.singular_values_[:20])
plt.plot(pca8.singular_values_[:20] ,'bo')
plt.xlabel("Number of Singular Values")
plt.ylabel("Singular Values")
plt.xticks(range(20))
plt.title("Decay of the singular values",fontsize = 16)
```

## Visualize two sets of results

```python
plt.figure(figsize=(18,6));

# Original Image
plt.subplot(1, 4, 1);
plt.imshow(PCA_data[0].reshape((218, 178, 3)),
#               cmap = plt.cm.gray, interpolation='nearest',
              clim=(0, 255));
plt.xlabel("All channels", fontsize = 12)
plt.title('Original Image: Prof.Paffenroth', fontsize = 13);




plt.subplot(1, 4, 2);
plt.imshow(approximation8[0].reshape((218, 178, 3)).astype(int),
#               cmap = plt.cm.gray, interpolation='nearest',
              clim=(0, 255));
plt.xlabel(str(n_components8) + ' components', fontsize = 12)
plt.title('80% of Explained Variance', fontsize = 13);


plt.subplot(1, 4, 3);
plt.imshow(approximation95[0].reshape((218, 178, 3)).astype(int),
#               cmap = plt.cm.gray, interpolation='nearest',
              clim=(0, 255));
plt.xlabel(str(n_components95) + ' components', fontsize = 12)
plt.title('95% of Explained Variance', fontsize = 13);



plt.subplot(1, 4, 4);
plt.imshow(approximation99[0].reshape((218, 178, 3)).astype(int),
#               cmap = plt.cm.gray, interpolation='nearest',
              clim=(0, 255));
plt.xlabel(str(n_components99) + ' components', fontsize = 12)
plt.title('99% of Explained Variance', fontsize = 13);
```
